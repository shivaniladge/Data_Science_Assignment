```python
nlp=spacy.load('en_core_web_sm')
nlp.pipe_names
```

```python
def calculate_sentiment(text):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score=sent_score+score.get(word.lemma_,0)
    return sent_score
```

```python
calculate_sentiment('Great')
```

```python
sent['emotions']=sent['Sentences'].apply(calculate_sentiment)
```

```python
sent
```

```python
#positive sentiment
sent[sent['emotions']>0]
```

```python
#negative sentiment
sent[sent['emotions']<0]
```

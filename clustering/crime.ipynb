```python
df.drop(columns='city',inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(df)
```

```python
x
```

```python
import scipy.cluster.hierarchy as hi
```

```python
lk=hi.linkage(x,method='complete')
```

```python
dendro=hi.dendrogram(lk)
plt.show()
```

```python
lk=hi.linkage(x,method='single')
dendro=hi.dendrogram(lk)
```

# KMEANS

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans()
```

```python
y=km.fit_predict(x)
```

```python
km.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,2],c=y,cmap='rainbow')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
sc=silhouette_score(x,y)
sc
```

```python
wcss =  []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
km=KMeans(n_clusters=4)
y=km.fit_predict(x)
```

```python
km.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'yellow',s=100)
```

```python
sc=silhouette_score(x,y)
```

```python
sc
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.5,min_samples=5)
```

```python
y=db.fit_predict(x)
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')
```

```python
from sklearn.metrics import silhouette_score
```

```python
db.labels_
```

```python
#k distance graph=hyperparameter tuning
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nb=NearestNeighbors(n_neighbors=3).fit(x)
```

```python
d,i=nb.kneighbors(x)
```

```python
pd.DataFrame(d)
```

```python
dist=np.sort(d[:,1])
```

```python
plt.plot(dist)
plt.xlabel('distance')
plt.ylabel('epsilon value')
```
